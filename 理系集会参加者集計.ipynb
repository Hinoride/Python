```python
# ファイル情報定義
LOG_FILE = "./20220121_理系集会/output_log_21-08-30.txt"
OUTPUT_FILE = "./output3.txt"
OUTPUT_FILE2 = "./output4.txt"
OUTPUT_FILE3 = "./output5.txt"

# 結果格納定義
joinList = []
leftList = []
tagList = []

# ログファイルを読み込み,必要な情報を結果変数に格納
with open(LOG_FILE, 'r', encoding="utf-8") as f:
    for row in f:
        if 'OnPlayerJoined' in row:
            joinList.append(row)
        elif 'OnPlayerLeft ' in row:
            leftList.append(row)
        elif '[Yodo][Dump]' in row:
            tagList.append(row)

#tagListから8桁の英数字をすべて抽出
import re
tagList3=[]
for tag in tagList:
    p = '00+[0-9,A-F]{6}'
    m = re.findall(p, tag)
    tagList3.append(m)

#抽出した8桁の英数字をDataFrameに変換し、ユニークな要素を抽出
import pandas as pd
import numpy as np
df=pd.DataFrame(tagList3)
df=df.replace({None:"00000000"})
df_2=np.unique(df)
print(df_2)

#DataFrameからリストに変換する
tagList4=[]
for o in df_2:
    tagList4.append(o)
#del tagList4[71:690]
print(len(tagList4))

#各8桁数字の要素数を各時刻ごとに計算し、リストに格納
tagList2 =[]
for u in tagList4:
    for tag in tagList:
        l=tag.count(str(u))
        tagList2.append(l)
tagList2=list(np.array_split(tagList2, len(tagList4)))

#8桁数字(16進数)を2進数に変換し、tagnumに格納
tagnum=[]
for t in range (len(tagList4)):
    q=bin(int(tagList4[t],16))
    tagnum.append(q)

#各8桁数字の要素数の最大値をtagmaxに格納
tagmax=[]
for i in range(len(tagList4)):
    y=np.amax(tagList2[i])
    tagmax.append(y)

#tagnumとtagmaxを辞書として結合
tag=dict(zip(tagnum, tagmax))

# 結果をファイルに書き込み(Joinのみ)
with open(OUTPUT_FILE, 'w', encoding="utf-8") as f:
    for row in joinList:
        f.write(row)

# 結果をファイルに書き込み(Leftのみ)
with open(OUTPUT_FILE2, 'w', encoding="utf-8") as f:
    for row in leftList:
        f.write(row)

# 結果をファイルに書き込み(tagのみ)
import csv
with open('tag.csv', 'w') as f:  
    writer = csv.writer(f)
    for k, v in tag.items():
       writer.writerow([k, v])

#総参加人数を計算
joinList2=[]
for i in range(len(joinList)):
    joinList2.append(joinList[i][61:])
joinList3=list(set(joinList2))
print("総参加人数" + str(len(joinList3))+"人")
with open(OUTPUT_FILE3, 'w', encoding="utf-8") as f:
    for row in joinList3:
        f.write(row)
```
